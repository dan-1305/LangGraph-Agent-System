## 1. Khởi tạo mã nguồn Streamlit (Lưu vào file app.py)

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

st.set_page_config(page_title="AI Stock Dashboard", layout="wide")
st.title("📈 AI Stock Forecasting Dashboard (Đồ án Môn học)")
st.markdown("**Thực hiện:** Phạm Công Danh & Nguyễn Công Tiến")

# Tạo 2 Tab Giao diện cho 2 phần riêng biệt
tab1, tab2 = st.tabs(["🤖 Học Máy (Machine Learning)", "🧠 Học Sâu (Deep Learning - LSTM)"])

with tab1:
    st.header("Dự báo Xu hướng Ngắn hạn (Phân loại 3 nhãn)")
    st.info("Sử dụng Random Forest và SVM để dự báo Tăng / Giảm / Đi ngang.")
    
    col1, col2 = st.columns(2)
    with col1:
        st.subheader("Kết quả Huấn luyện")
        st.write("- **Random Forest Accuracy:** 47.45%")
        st.write("- **SVM Accuracy:** 51.76%")
        st.warning("Do tính ngẫu nhiên (Random Walk) của thị trường, các chỉ báo kỹ thuật cơ bản gặp khó khăn khi phân loại 3 nhãn.")
    
    with col2:
        st.subheader("Tầm quan trọng của Đặc trưng (Feature Importance)")
        img_path_ml = '../01_Data/ml_feature_importance.png'
        if os.path.exists(img_path_ml):
            st.image(img_path_ml, caption="Mô hình Random Forest đánh giá mức độ quan trọng")
        else:
            st.write("*(Đang chờ nạp biểu đồ...)*")

with tab2:
    st.header("Dự báo Chuỗi Thời Gian - 5 Phiên Kế Tiếp")
    st.info("Sử dụng mạng LSTM 2 Lớp với PyTorch Backend (Trượt cửa sổ 60 ngày).")
    
    col3, col4 = st.columns(2)
    with col3:
        st.subheader("Biểu đồ Hàm Mất mát (Loss)")
        img_path_loss = '../01_Data/lstm_loss_chart.png'
        if os.path.exists(img_path_loss):
            st.image(img_path_loss, caption="Kiểm soát Overfitting qua các Epochs")
        else:
            st.write("*(Đang chờ nạp biểu đồ...)*")
            
    with col4:
        st.subheader("Kết quả Dự báo (Multi-step)")
        img_path_pred = '../01_Data/lstm_prediction_chart.png'
        if os.path.exists(img_path_pred):
            st.image(img_path_pred, caption="Thực tế vs Dự đoán")
        else:
            st.write("*(Đang chờ nạp biểu đồ...)*")
```
